```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Titanic dataset (you can download it from Kaggle or use any dataset)
df = pd.read_csv('train.csv')
```

```python
df.info
```

```python
df.isnull().sum()
```

```python
# 1. Data Cleaning (Handle missing values)
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# 2. Exploratory Data Analysis (EDA)
# Plot survival rate based on gender
#survived (1) versus those who didn't survive (0).
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Sex', data=df, palette='coolwarm')
plt.title('Survival Rate Based on Gender')
plt.show()
```

```python
# Plot age distribution of passengers
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='blue')
plt.title('Age Distribution of Passengers')
plt.show()

# 3. Explore relationships (e.g., between Pclass and Survival)
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='coolwarm')
plt.title('Survival Rate Based on Passenger Class')
plt.show()
```
